```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def create_tree(lista):
    tam = len(lista)-1
    for i in range(tam):
        if 2*i+1 <= tam:
            G.add_edge(lista[i],lista[2*i+1])
        if 2*i+2 <= tam:
            G.add_edge(lista[i],lista[2*i+2])
```

```python
G = nx.Graph()
lista = range(30)
G.add_nodes_from(lista)
create_tree(lista)
nx.draw_kamada_kawai(G,node_color = 'lime',node_size=500,with_labels = True)
# nx.draw_networkx(G,node_color = 'lime',node_size=1600,with_labels = True)
# limits=plt.axis('off')
```
