```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
# adicionar v√©rtices
G.add_node('exemplo')
G.add_nodes_from(range(1,3))

# adicionar arestas
G.add_edge(1,'exemplo')
G.add_edges_from([(1, 2),(1,3),(3,1),(2,1),('exemplo',1)])
```

```python
nx.draw_networkx(G,node_color = 'b',node_size=1600,with_labels = True)
limits=plt.axis('off') # turn of axis
```

```python
G = nx.petersen_graph()
nx.draw_networkx(G,node_color = 'b',node_size=1600,with_labels = True)
limits=plt.axis('off') # turn of axis
```
