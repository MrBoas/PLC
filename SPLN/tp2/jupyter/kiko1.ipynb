```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.petersen_graph()

plt.subplot(231)
nx.draw_shell(G, with_labels=True, font_weight='bold')
plt.subplot(232)
nx.draw_spectral(G, with_labels=True, font_weight='bold')
plt.subplot(236)
nx.draw_shell(G, with_labels=True, font_weight='bold')
plt.savefig("path.png", format="png") # nao pode ter show antes
# plt.close('all') # fecha a janela, kinda
```
