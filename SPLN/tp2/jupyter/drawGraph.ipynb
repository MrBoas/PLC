```python
import networkx as nx
import matplotlib.pyplot as plt

triplos = [('Harry', 'Potter', 21), ('Harry', 'Luna', 22), ('Harry', 'Neville', 22), ('Percy', 'Harry', 25), ('Harry', 'Jorge', 30), ('Harry', 'Fred', 33), ('Fred', 'Jorge', 48), ('Harry', 'VÃ¡lter', 67)]
```

```python
def get_nodes(triplos):
    set1 = set([t1 for t1,t2,n in triplos])
    set1.update([t2 for t1,t2,n in triplos])
    # print(set1) # debug
    return set1
```

```python
def draw(nodes,edgesW):
    G = nx.Graph()
    G.add_nodes_from(nodes)
    G.add_weighted_edges_from(edgesW)
    nodeSize = [len(node)*200+200 for node in nodes]
    # print(nodes) # debug
    # print(nodeSize) # debug

    # pos = nx.shell_layout(G) # alternative layout
    pos = nx.circular_layout(G)
    nx.draw(G,pos,nodelist=nodes,node_size=nodeSize,with_labels=True)
    labels = nx.get_edge_attributes(G,'weight')
    nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
    plt.savefig("tp2.png")
```

```python
nodes = get_nodes(triplos)
edgesW = triplos
draw(nodes,edgesW)
```
